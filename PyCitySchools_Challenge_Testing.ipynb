```python
prefixes_and_suffixes = [' MD', ' DVM', ' DDS', ' PhD', 'Dr. ', 'Mr. ', 'Miss ', 'Mrs. ', 'Ms. ']
```

```python
for word in prefixes_and_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
student_data_df['student_name']
```

```python
student_list=student_data_df["student_name"].tolist()
students_to_fix =[]
suffix = []
prefix = []
for student in student_list:
    student_name = student.split()
    if len(student_name) >= 3:
        students_to_fix.append(student)
students_to_fix
for student in students_to_fix:
    student_name = student.split()
    if student_name[-1] not in suffix:
        suffix.append(student_name[-1])
    if student_name[0] not in prefix:
        prefix.append(student_name[0])
print(suffix)
print(prefix)
```

```python
passing_math_df = school_data_complete_df[school_data_complete_df['math_score'] >= 70]
passing_reading_df = school_data_complete_df[school_data_complete_df['reading_score'] >= 70]
students_passing_math = passing_math_df['math_score'].count()
students_passing_reading = passing_reading_df['reading_score'].count()
print(students_passing_math)
print(students_passing_reading)
```
