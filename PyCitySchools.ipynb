```python
prefixes_and_suffixes = [' MD', ' DVM', ' DDS', ' PhD', 'Dr. ', 'Mr. ', 'Miss ', 'Mrs. ', 'Ms. ']
```

```python
for word in prefixes_and_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
student_data_df['student_name']
```

```python
student_data_df.head()
```

```python
student_list=student_data_df["student_name"].tolist()
students_to_fix =[]
suffix = []
prefix = []
for student in student_list:
    student_name = student.split()
    if len(student_name) >= 3:
        students_to_fix.append(student)
students_to_fix
for student in students_to_fix:
    student_name = student.split()
    if student_name[-1] not in suffix:
        suffix.append(student_name[-1])
    if student_name[0] not in prefix:
        prefix.append(student_name[0])
print(suffix)
print(prefix)
```

```python
school_data_complete_df = pd.merge(clean_student_data_df,school_data_df,on=['school_name','school_name'])
school_data_complete_df.head()
```

```python
student_count = school_data_complete_df['student_name'].count()
student_count
```

```python
schools_count = len(school_data_complete_df['school_name'].unique())
schools_count
```

```python
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
mean_math_total = school_data_complete_df['math_score'].mean()
mean_reading_total = school_data_complete_df['reading_score'].mean()
print(mean_math_total)
print(mean_reading_total)
```

```python
passing_math_df = school_data_complete_df[school_data_complete_df['math_score'] >= 70]
passing_reading_df = school_data_complete_df[school_data_complete_df['reading_score'] >= 70]
students_passing_math = passing_math_df['math_score'].count()
students_passing_reading = passing_reading_df['reading_score'].count()
print(students_passing_math)
print(students_passing_reading)
```

```python
percent_pass_math = passing_math_df['Student ID'].count() / school_data_complete_df['Student ID'].count() *100
percent_pass_reading = passing_reading_df ['Student ID'].count() / school_data_complete_df['Student ID'].count() *100
print(percent_pass_math)
print(percent_pass_reading)
```

```python
pass_math_reading_df = school_data_complete_df[(school_data_complete_df['math_score']>=70) & 
                                          (school_data_complete_df['reading_score']>=70)]
percent_pass_all = pass_math_reading_df['Student ID'].count() / school_data_complete_df['Student ID'].count() *100
print(percent_pass_all)
```

```python
district_summary_df = pd.DataFrame([{'Total Schools':schools_count,
                                   'Total Students':student_count,
                                   'Total Budget': total_budget,
                                   'Average Reading Score':mean_reading_total,
                                   'Average Math Score': mean_math_total,
                                   '% Passing Reading': percent_pass_reading,
                                   '% Passing Math': percent_pass_math,
                                   '% Overall Passing': percent_pass_all}])
district_summary_df
district_summary_df.dtypes
```

```python
district_summary_df['Total Schools'] = district_summary_df['Total Schools'].map("{:,}".format)
district_summary_df['Total Students'] = district_summary_df['Total Students'].map("{:,}".format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map("${:,.2f}".format)
district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map("{:.1f}".format)
district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map("{:.1f}".format)
district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map("{:.0f}".format)
district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map("{:.0f}".format)
district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map("{:.0f}".format)
district_summary_df
```

```python
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", 
                    "% Passing Math", "% Passing Reading", "% Overall Passing"]
district_summary_data_reorder_df = district_summary_df[new_column_order].copy()
```
